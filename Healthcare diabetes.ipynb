# Correlation Analysis

```python
data.corr()
```

```python
plt.figure(dpi=80)
sns.heatmap(data.corr(),cmap='viridis')
```

```python
data
```

```python


x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=40,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=25) 
#Using 25 Neighbors just as thumb rule sqrt of observation
knn_model.fit(x_train,y_train)
knn_pred=knn_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve,auc
```

```python
print("Accuracy Score :",accuracy_score(knn_pred,y_test))
print("Conusion Matrix \n:",confusion_matrix(knn_pred,y_test))
print("classification_report \n:",classification_report(knn_pred,y_test))
```

```python
probs=knn_model.predict_proba(x_test)
```

```python
probs=probs[:,1]
```

```python
roc_auc_score(y_test,probs)
```

```python
fpr,tpr,thresholds=roc_curve(y_test,probs)
roc_auc_knn=auc(fpr,tpr)
```

```python
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr,tpr,'b',label='AUC Score = %0.2f'%roc_auc_knn)
plt.plot(fpr,fpr,'r--',color='red')
plt.legend()
```

```python
from sklearn.svm import SVC
svc_model_linear = SVC(kernel='linear',random_state=0,probability=True,C=0.01) 
svc_model_linear.fit(x_train,y_train)
svc_pred=svc_model_linear.predict(x_test)
```

```python
print("Accuracy Score :",accuracy_score(svc_pred,y_test))
print("Conusion Matrix \n:",confusion_matrix(svc_pred,y_test))
print("classification_report \n:",classification_report(svc_pred,y_test))
```

```python
probs_svc=svc_model_linear .predict_proba(x_test)
```

```python
probs_svc=probs_svc[:,1]
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(C=0.01) 
lr_model.fit(x_train,y_train)
lr_pred=lr_model.predict(x_test)
print("Accuracy Score :",accuracy_score(lr_pred,y_test))
print("Conusion Matrix \n:",confusion_matrix(lr_pred,y_test))
print("classification_report \n:",classification_report(lr_pred,y_test))
```
